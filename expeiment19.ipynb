```python
import pandas as pd
import plotly.express as px
df=pd.DataFrame({
    'Department' : ['HR','IT','SALES','MARKETING'],
    'Budget': [100, 150, 120, 80] # Added 'Budget' column with example values
})
fig=px.treemap(df,path=['Department'],values='Budget')
fig.show()
```

```python
#2.Dendogram
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
#Sample data (can represent students,features,etc.)
data=np.array([
    [5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30]
])

#perform hierarchial clustering
linked=linkage(data,method='ward')

#plot the dendrogram
plt.figure(figsize=(6,4))
dendrogram(linked)
plt.title('Rishabh')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
#3.Venn Diagram
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
#Define sets
A=set([1,2,3,4]) # Convert list to set
B=set([3,4,5,6]) # Convert list to set
#Create Venn Diagram
venn2([A,B],set_labels=('Set A','Set B'))
#Title
plt.title('Rishabh')
plt.show()
```

```python
#Sankey Diagram(Student Flow)
import plotly.graph_objects as go
fig=go.Figure(data=[go.Sankey(
    node=dict(label=["Admission","First Year","Second Year","Placed"]),
    link=dict (
        source=[0,1,2],
        target=[1,2,3],
        value=[100,80,60]
    )
)])
fig.show()
```

```python
#5.3D scatter plot(Student Perfomance)
import plotly.express as px
df=pd.DataFrame({
    'Study Hours': [2,4,6,8,5],
    'Marks': [50,65,75,90,70],
    'Attendance':[60,75,80,90,70]
})
fig=px.scatter_3d(df,x='Study Hours',y='Marks',z='Attendance',title='Anurag')
fig.show()
```

```python
#6. Radar Chart(Skill assessment)
import plotly.graph_objects as go
skills = ['Python','ML','DBMS','DSA','Communication']
values = [4, 3, 5, 4, 3]
fig = go.Figure(go.Scatterpolar(
    r=values,
    theta=skills,
   fill='toself',
   name='Student Skills'
))
fig.show()
```
